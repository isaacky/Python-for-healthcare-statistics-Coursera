## Computer variables

**Computer variables** are names that we give to a space in memory, where we want to save an object.  Just to mention something that will not make a lot of sense now, but we can give different names to the same space in memory.

Everything (we called them *things* before) that we have seen so far are **objects**.  This includes simple numbers, strings, the solution to calculations.  We have seen the `type()` function that gives us the type of an object.

There are some restrictions to the names we can use for computer variables:

- Use names that start with a lowercase letter
- Make it descriptive
- Use snake case `my_variable_name` (or the less popular camel case `myNewVariable`)
- Don't use `python` functions

- Creating a computer variables using the assignment operator

```
# Create a computer variable to hold a floating point object
my_value = 3.0  # Assign an object which is a  floating point number to a computer variable using the assignment operator
```

```
# Recall the object in the computer variable
my_value
```

```
# Type of object held in the computer variable
type(my_value)
```

- Assignment update

```
# Starting value
i = 0
i
```

```
# Add 1 to i
i = i + 1
i
```

```
# Short-hand update
i += 1
i
```

## Dictionaries

**Dictionaries** are key-value pairs.  Every value has a key (a name).  Python uses curly braces to indicate dictionaries.

```
# create a dictionary of key-value pairs
my_dict = {'Language':'Python',
           'Version':3,
           'Environment':'Notebook'}
```

We can call up the keys with the method `.keys()` and the values with the method `.values()`.

```
# Keys
my_dict.keys()
```

```
# Values
my_dict.values()
```

All the key-value pairs can be retrieves using the `.items()` method.

```
# Retrieve keys and values
my_dict.items()
```

To het the value of a specific key, we use the `.get()` method.

```
# Value of Version key
my_dict.get('Version')
```

The alternative syntax uses a list of tuples.

```
# Alternative syntax with list of tuples
my_other_dict = dict([('Language', 'Python'),
                      ('Version', 3.8),
                      ('Environment', ['Spyder', 'Notebook'])])
my_other_dict
```

## Loops

Loops allows us to run over some code many times.

- `for` loop

```
# Loop through 0 to 9
for i in range(10):
  print(i)
```

We can loop through the elements of a list object.  The `element` name down below is just a placeholder name.

```
my_other_list
```

```
# Loop through elemnts in my_other_list and print them
for element in my_other_list:
  print(element)
```

```
# Loop through keys in my_dict and print them
for key in my_other_dict.keys():
  print(key)
```

We can also loop over code while a condition is met.

- `while` loop

```
# Print values while less than 5
i = 0  # Start value
while i < 5:
  print(i)
  i += 1
```

## Conditionals

Sometimes, we only want to execute code when some condition is met.  The `if`, `elif`, and `else` statements all help us do this.

Below we store an integer object in a computer variable.  When then set a condition.  If the intege object's value is less than a certain number, we print something to the screen, but if it is not, then we print something else to the screen.

```
# Check on condition
i = 23
if i < 20:
  print('Smaller than 20')
else:
  print('Equal to or greater than 20')
```

```
# Using elif
a = 2
b = 1
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
```

### Example excerise

In this famous computer task, we take a range of numbers (0 through 21 in this case).  For each number, if it is divisible by some integer, say 3 (no remainder), then we print `foo`, else we print `bar`.

```
# foo-bar
for i in range(22):
  if i % 3 == 0:
    print(i, ' : foo')
  else:
    print(i, ' : bar')
```

## List comprehension

This is a quick and easy way to create a list.

- Integers 0 through 50

```
my_set = [i for i in range(51)]
my_set[0:5]  # Print first 5
```

- Show every number in above list that is divisible by 3

```
[i for i in my_set if i % 3 == 0]
```

- Square every number in the list that is divisible by 5

```
[i**2 for i in my_set if i % 5 == 0]
```
